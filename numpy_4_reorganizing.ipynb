## Reorganizing Arrays

```python
before = np.array([[1,2,3,4],[5,6,7,8]])
print(before)
```

```python
after = before.reshape((4,2))
print(after)
```

If the values amounts dont much, it will rase type error

```python
# Vertically stacking vectors?
v1 = np.array([1,2,3,4])
v2 = np.array([5,6,7,8])
np.vstack([v1,v2,v1,v2])
```

```python
# Horizontal stack
h1 = np.zeros((2,4))
h2 = np.ones((2,2))
np.hstack((h1,h2))
```
