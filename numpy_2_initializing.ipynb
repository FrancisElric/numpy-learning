````python
import numpy as np
````

## Initializing Different Types of Arrays

````python
# All 0s matrix
print(np.zeros((2,2)))
print(np.zeros((2,2,2),dtype='int8'))
````

````python
# All 1s matrix
print(np.ones((2,2)))
````

````python
# Any other number
np.full((2,2), 99, dtype='float16')
````

````python
# Any other number with full_like (not that usefull?)
a = np.array([[9.,8.,7.],[6.,5.,4.]])
np.full_like(a, 4)
````

````python
# Random decimal numbers
print(np.random.rand(4,2))
print('####')
print(np.random.random_sample(a.shape))
````

````python
# Random Int values
np.random.randint(-8,8, size=(3,3))
````

````python
arr = np.array([[1,2,3]])
r1 = np.repeat(arr, 3, axis=0)
print(r1)
````

Be careful when copying arrays!

````python
a = np.array([1,2,3])
b = a
b[0] = 100
print(a)
````

b = a makes b point to the SAME array that a is, to prevent it, use .copy()

````python
a = np.array([1,2,3])
b = a.copy()
b[0] = 100
print(a)
````
